importing necessary libraries  

```python
from PIL import Image
import pytesseract
import os
from google.colab.patches import cv2_imshow
import cv2
```

function to extract the text from the image using pytesseract

```python
def OCR(image_path):
  img=Image.open(image_path)
  image=cv2.imread(image_path)
  cv2_imshow(image)
  string=pytesseract.image_to_string(img, lang='eng')
  print(string)
```

Extracting text from printed image1

```python
OCR('test1.jpg')
```

```python
OCR('test2.jpg')
```
